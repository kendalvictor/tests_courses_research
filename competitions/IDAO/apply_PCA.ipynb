```python
train.columns, train.shape
```

```python
def get_cols_similary(val):
     return [col for col in train.columns if col.startswith(val)]
```

```python
dicc_pca = {
    'ncl': {'columns': get_cols_similary('ncl')},
    'avg_cs': {'columns': get_cols_similary('avg_cs')},
    'MatchedHit_X': {'columns': get_cols_similary('MatchedHit_X')},
    'MatchedHit_Y': {'columns': get_cols_similary('MatchedHit_Y')},
    'MatchedHit_Z': {'columns': get_cols_similary('MatchedHit_Z')},
    'MatchedHit_DX': {'columns': get_cols_similary('MatchedHit_DX')},
    'MatchedHit_DY': {'columns': get_cols_similary('MatchedHit_DY')},
    'MatchedHit_DZ': {'columns': get_cols_similary('MatchedHit_DZ')},
    'MatchedHit_T[': {'columns': get_cols_similary('MatchedHit_T[')},
    'MatchedHit_DT': {'columns': get_cols_similary('MatchedHit_DT')},
    'Lextra_X': {'columns': get_cols_similary('Lextra_X')},
    'Lextra_Y': {'columns': get_cols_similary('Lextra_Y')},
    'Mextra_DX2': {'columns': get_cols_similary('Mextra_DX2')},
    'Mextra_DY2': {'columns': get_cols_similary('Mextra_DY2')},
}
```

```python
columns_type = get_cols_similary('MatchedHit_TYPE')
print(columns_type)
```

```python
corr_spearman_type = train[columns_type].corr(method='spearman')
corr_detail(corr_spearman_type, min_value=0.75, show_detail=True)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
for k, v in dicc_pca.items():
    data_train = train[v.get('columns', [])]
    data_test = test[v.get('columns', [])]
    
    escala = StandardScaler()
    escala.fit(data_train)

    escalada_train =  pd.DataFrame(escala.transform(data_train))
    escalada_test =  pd.DataFrame(escala.transform(data_test))
    del data_train
    del data_test
    
    pca = PCA(n_components=2)
    pca.fit(escalada_train)
    
    col_1 = k + '_1'
    col_2 = k + '_2'
    v['cols_train'] = pd.DataFrame(pca.transform(escalada_train), columns=[col_1, col_2])
    v['cols_test'] = pd.DataFrame(pca.transform(escalada_test), columns=[col_1, col_2])
    del escalada_train
    del escalada_test
```

```python
for k, v in dicc_pca.items():
    cols = v.get('columns', [])
    for col in cols:
        del train[col]
        del test[col]
        
    train = pd.concat([train, v.get('cols_train')], axis=1)
    test = pd.concat([test, v.get('cols_test')], axis=1)    
```
