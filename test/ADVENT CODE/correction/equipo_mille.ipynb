```python
# coding: utf-8

# In[18]:


import itertools


# In[11]:


lines = [line.rstrip('\n') for line in open("rutas.txt")]
lines


# In[12]:


cities = {}


# In[13]:


indice = 0


# In[26]:


grafo = []
for i in range(8):
    grafo.append([])
for i in range(8):
    for j in range(8):
        grafo[i].append(0)


# In[28]:


for line in lines:
    data = line.split()
    if data[0] not in cities:
        cities[data[0]] = indice
        indice += 1
    if data[2] not in cities:
        cities[data[2]] = indice
        indice += 1
    grafo[cities[data[0]]][cities[data[2]]] = int(data[4])
    grafo[cities[data[2]]][cities[data[0]]] = int(data[4])


# In[29]:


cities


# In[30]:


permutaciones = list(itertools.permutations(range(8))) ###modificar numero de cuidades


# In[31]:


meta = int(1000000000)
for perm in permutaciones:
    currentDistance = 0
    ruta = list(perm)
    for i in range(1,len(ruta)):
        currentDistance += grafo[ruta[i]][ruta[i-1]]
    if(currentDistance<meta):
        meta = currentDistance


# In[32]:


meta


# In[33]:


grafo
```
