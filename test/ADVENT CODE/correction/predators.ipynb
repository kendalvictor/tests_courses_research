```python
#! /usr/bin/python3 -s

# -*- coding: utf-8 -*-
"""
Spyder Editor

@autor: Jahir M.

This is a temporary script file.
"""

from __future__ import with_statement

from copy import copy as copy
from os.path import getsize as getsize
from math import ceil, sqrt

from time import time as time

def exec_clock(method):

    def timed(*args, **kw):
        start= time()
        result = method(*args, **kw)
        end = time()
        
        val_diff = end - start
        print("""
        Tiempo: \t{:.10f} Seg | Ratio: \t{:.4f}% de un seg
        """.format(val_diff, val_diff * 100))
        return result

    return timed

def cond_adde(dict_ref, id_x, key):
    try:
        temp = dict_ref[key]
    except:
        dict_ref[key] = id_x
        return [id_x, id_x + 1]
    else:
        return [temp, id_x]

@exec_clock        
def get_data(FILE_PATH):
    """
    Precalculando el numero de elementos para separar memoria:
    
    Dado que se espera un "mapa de nodos" , y las distancias siempre estan dadas
    en pares, es una combinatiora de n elementos de 2 en 2, teniendo como formula
    
    comb = (n^2 - n) / 2
    0 = n^2 - n - 2*comb
    > https://i.imgur.com/CSkPTJg.png prueba del comportamiento
    cuadratico
    
    por formula general:
    
    roots = (-b +- sqrt(b^2 - 4ac))/ 2a
    aprox_elemmt = 1 +- sqrt(1 + 4*2*comb))/ 2
    
    como el minimo numero de combinaciones es 1 y los elementos siempre son
    positivos, solo se debe usar la raiz positiva (+- => +)
    
    aprox_elemnt = ceil(0.5 + sqrt(1 + 8*comb)/2)
    
    """
    
    _file_size = getsize(FILE_PATH)
    with open(FILE_PATH) as f:
        _first_line = f.readline()
    _line_size = len(_first_line.encode('utf-8'))
    
    aprox_comb = ceil((_file_size * 1.0)/ _line_size)
    aprox_elemnt = ceil(0.5 + sqrt(1 + 8*aprox_comb)/2)
    

    #print("""
    #(en bytes)
    #Tamano por linea : {:.0f}
    #Tamano de archivo : {:.0f}
    #numero de combinaciones (numero de lineas, aprox.) : {:.0f}
    #Numero de ciudades (elementos, aprox.) : {:.0f}
    #""".format(_line_size, _file_size, aprox_comb, aprox_elemnt))

    
    """
    calculado el numero de elementos aproximado, se procede a separar memoria.
    
    Por que no se separa memoria como en C/C++? (asumiendo esta pregunta es
    del tipo programacion competitiva)
    
    Por que la forma python-ica de separar memoria en python no es tan eficiente
    como en C/C++, ademas de que en C/C++ se asume existen limites de memoria
    por cada ejercicio. Mientras en esta pregunta se esta asumiendo una expectativa
    de generalizacion alta.
    """
    mtx_ady = [ [None] * ll + [0] for ll in range(int(aprox_elemnt))]
    
    #print("\nMatriz de Adyacencia (Vacia)\t:\n\n")
    #print(mtx_ady)
    
    """
    Intercambio de uso: +memoria -iteraciones => +memoria -uso de cpu
    
    el diccionario servira para tener asociado los indices de la futura matriz de 
    adyacencia a los nombres de sus ciudades
    
    y la matriz de adyacencia se almacena solo la mitad diagonal (superior o 
    inferior) , dado que las rutas son bidireccionales
    """
    
    file = open(FILE_PATH, "r")
    
    city_dict = {}
    
    
    gen_id_x = 0
    for ii in file:
        [cityA, cityB, val] = ii.rstrip().replace(" to ", "=").replace(" ", "").split("=")
        
        [id_i, gen_id_x] = cond_adde(city_dict, gen_id_x, cityA)
        [id_j, gen_id_x] = cond_adde(city_dict, gen_id_x, cityB)
        
        mtx_ady[id_j][id_i] = int(val)
    
    #print("\nCiudades :\t")
    #print(gen_id_x)
    #print("\nMatriz de Adyacencia (Llena)\t:\n\n")
    #print(mtx_ady)
    
    return mtx_ady, city_dict


### midiendo eficiencia
if __name__ == '__main__':
    #import timeit
    #print(timeit.timeit("get_data(mtx_ady, city_dict)", 
    #                   setup="from __main__ import get_data"))
    
    FILE_PATH = "./rutas.txt"    
    
    mtx_ady, city_dict = get_data(FILE_PATH)
    
    print(mtx_ady, city_dict)
    



"""
Calculando distancias por posibles recorridos:

dado que se pide la distancia minima, y asumiendo en todo el procedimiento 
anteror las ciudades forma un grafo simple conexo no ponderado 
sin sumideros o fuentes; pudiendo asi omitir :
    - el aliasing de nodos (dos ciudades que siempre tienen 
    distancia 0, por tanto son la misma)
    - trabajar con un formato flujo minimo, alcance maximo (modificado)
    
Sin embargo no hay origen definido, por lo que esta situacion se convierte en
el peor caso del problema del viajero (sin retorno) , el caso donde se debe 
hallar la ruta mas corta global
"""

### Algoritmo Genetico


    




    
```
