```python
# -*- coding: utf-8 -*-
"""
Created on Wed Apr 24 14:44:05 2019

@author: jce1
"""

"""CARGAMOS LAS LIBRERIAS NECESARIAS"""

import pandas as pd
import numpy as np

"""CARGAMOS LOS DATOS"""

rutas = pd.read_table("rutas.txt", sep = ' ', header = None)

"""Eliminamos las columnas que no se utilizaran"""
rutas = rutas.iloc[:,[0,2,4]]

"""Nombramos las columnas"""
rutas.columns = ['E1', 'E2','D']

"""Listamos los lugares"""
place = list(rutas.E1.unique())

"""Creamos funciones utiles para las permutaciones de 7"""
def inserta(x, lst, i):
    """Devuelve una nueva lista resultado de insertar
       x dentro de lst en la posición i.
    """
    return lst[:i] + [x] + lst[i:]

def inserta_multiple(x, lst):
    """Devuelve una lista con el resultado de
       insertar x en todas las posiciones de lst.  
    """
    return [inserta(x, lst, i) for i in range(len(lst) + 1)]

def permuta(c):
    """Calcula y devuelve una lista con todas las
       permutaciones posibles que se pueden hacer
       con los elementos contenidos en c.
    """
    if len(c) == 0:
        return [[]]
    return sum([inserta_multiple(c[0], s)
                for s in permuta(c[1:])],
               [])
    
def imprime_ordenado(c):
    """Imprime todos los subconjuntos del conjunto 
       c (una lista de listas) ordenados. Cada subconjunto
       se imprime en su propia línea.
    """
    aux = []
    
    for e in sorted(c, key=lambda s: (len(s), s)):
        aux.append(e)
        
    return aux

"""Creamos funciones utiles para las permutaciones de 7"""
data = imprime_ordenado(permuta(place))
data = pd.DataFrame(data)
data.columns = ['P1','P2','P3','P4','P5','P6','P7']
data['Dist'] = np.nan

"""Renombramos los datasets"""
full = data
data = rutas

"""Generando la distancia para todas las permutaciones"""
for i in range(len(full)):
    
    suma = 0
    
    for j in range(6):
        
        a1 = full.iloc[i,j]
        a2 = full.iloc[i,j+1]
        
        for k in range(len(data)):
            
            if (a1 == data.loc[k,'E1']) & (a2 == data.loc[k,'E2']):
                
                suma = suma + data.loc[k,'D']
                break
            
            elif (a2 == data.loc[k,'E1']) & (a1 == data.loc[k,'E2']):
                
                suma = suma + data.loc[k,'D']
                break
                
    full.loc[i,'Dist'] = suma
        
del i, j, k, suma, a1, a2


"""Resultado final : Distancia mínima"""
distancia_minima = min(full.Dist)
distancia_minima

"""Resultado final : Combinacion mínima"""
combinacion = full[ full.Dist == full.Dist.min()]
combinacion
```
