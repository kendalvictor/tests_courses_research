```python
import collections
import re
import numpy as np
"""
Definicion del grafo
"""
class Graph:
    def __init__(self):
        self.nodes = set()
        self.edges = collections.defaultdict(list)
        self.distances = {}
    """
    Para agregar un nodo
    """
    def add_node(self, value):
        self.nodes.add(value)
    """
    Para agregar un borde
    """
    def add_edge(self, from_node, to_node, distance):
        #grafo no dirijido
        self.edges[from_node].append(to_node)
        self.edges[to_node].append(from_node)
        self.distances[(from_node, to_node)] = distance
        self.distances[(to_node, from_node)] = distance
    """
    Para imprimir todo el grafo
    """
    def __str__(self):
        res = "nodes: "
        for k in self.nodes:
            res += str(k) + " "
        
        res += "edges: "
        res += str(self.distances)
        
        return res

def obtainPath(index2node, i, j, parents, distances):
    if distances[i][j] == 9999:
        return " no path to "
    if parents[i][j] == i:
        return "->"
    else :
        return obtainPath(index2node, i, parents[i][j], parents, distances) +index2node[parents[i][j]] + obtainPath(index2node, parents[i][j], j, parents, distances)
    
def main():
    with open("rutas.txt", "r") as file:    
        all_routes = file.read()
    all_routes = all_routes.split('\n') #lista con todas las conexiones posibles
    all_nodes = []
    
    g = Graph() #se crea el grafo vacio
    
    for r in all_routes:
        #procesamiento de la infomarcion
        temp = re.findall(r'(.*?) to (.*?) = (\d+)', r)[0]
        from_node = temp[0]
        all_nodes.append(from_node)
        to_node = temp[1]
        distance = int(temp[2])

        #agregamos los nodos
        g.add_node(from_node)
        g.add_node(to_node)
        #agregamos la ruta
        g.add_edge(from_node, to_node, distance)
        
    size_vertex = len(g.nodes)
    distances = np.full((size_vertex, size_vertex), 9999)
    for i in range (0,size_vertex):
        for j in range (0,size_vertex):
            if i == j:
                distances[i][j] = 0

    parents = np.full((size_vertex, size_vertex), 0)
    for i in range (0,size_vertex):
        for j in range (0,size_vertex):
            if distances[i][j] == 9999:
                parents[i][j] = 0
            else:
                parents[i][j] = i
    node2index = {e:i for i,e in enumerate(g.nodes)}
    index2node = {i:e for i,e in enumerate(g.nodes)}
    for edge, distance in g.distances.items():
        distances[node2index[edge[0]]][node2index[edge[1]]] = distance
        distances[node2index[edge[1]]][node2index[edge[0]]] = distance
        parents[node2index[edge[0]]][node2index[edge[1]]] = node2index[edge[0]]
        parents[node2index[edge[1]]][node2index[edge[0]]] = node2index[edge[1]]

    for k in range(size_vertex):  
        for i in range(size_vertex): 
  
            # Se prueba todos los vertices como destino desde el punto de
            # partida k
            for j in range(size_vertex): 
  
                # Se realiza una actualizacion de las distancias 
                # si la distance es menor que la sumatoria
                if distances[i][j] > distances[i][k]+ distances[k][j]: 
                    distances[i][j] = distances[i][k]+ distances[k][j] 
                    parents[i][j] = parents[k][j]
    for i in range(size_vertex):
        for j in range(i+1, size_vertex):
            print('Ruta:', index2node[i], obtainPath(index2node, i,j, parents, distances), index2node[j], '|| Distancia:', distances[i][j])          
```

```python
main()
```
