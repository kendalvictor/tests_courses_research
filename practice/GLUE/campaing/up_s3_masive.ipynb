```python
from datetime import datetime, timedelta
import boto3
import glob
import json
import os
```

```python
client = boto3.client('s3')
s3 = boto3.resource('s3')
```

```python
AWS_DEFAULT_ACL = 'public-read'
BUCKET_NAME = 'aatempjson2'
REGION = 'us-east-1'

bucket = s3.Bucket(BUCKET_NAME)
if not bucket.creation_date:
    try:
        s3.create_bucket(
            ACL=AWS_DEFAULT_ACL,
            Bucket=BUCKET_NAME
        )
    except Exception as e:
        print('-> ', str(e))
```

```python
%%time
marca = 'elcomercio'
start = datetime.strptime("2019-03-01", "%Y-%m-%d").date()
end = datetime.strptime("2019-03-05", "%Y-%m-%d").date()
campains = ['saludos', 'despedidas', 'reuniones']

for dayy in (start + timedelta(days=_) for _ in range(0, (end - start).days + 1)):
    dayy = str(dayy)
    for campain in campains:
        for i in range(1000):
            dicc = {
                'id': i,
                'metric_1': 1,
                'metric_2': 2,
                'metric_3': 3,
                'metric_4': 4,
            }
            path = '{}/{}/{}/send-{}.json'.format(marca, dayy, campain, i)
            s3.Object(BUCKET_NAME, path).put(
                Body=json.dumps(dicc), ACL='public-read'
            )
            print(path)
```
