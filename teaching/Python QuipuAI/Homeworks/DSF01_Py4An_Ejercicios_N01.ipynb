## Ejercicio 4 (6 puntos)

En base al ejercicio anterior, aplicar la función lambda sobre una *lista* de diccionarios que contiene el nombre y monto prestado a una lista de personas.

lista_diccionarios = [ {'id': 1, 'nombre': 'Alberto', 'prestamo': 10000},

{'id': 2, 'nombre': 'Freddy', 'prestamo': 8000},

{'id': 3, 'nombre': 'Yomira', 'prestamo': 20000},

{'id': 4, 'nombre': 'Juan', 'prestamo': 6000}]

Y obtener por cada persona el monto total a pagar.

### Ejemplo 

**Input:**

lista_diccionarios = [ {'id': 1, 'nombre': 'Alberto', 'prestamo': 10000},

{'id': 2, 'nombre': 'Freddy', 'prestamo': 8000},

{'id': 3, 'nombre': 'Yomira', 'prestamo': 20000},

{'id': 4, 'nombre': 'Juan', 'prestamo': 6000}]

Tasa de interés total: 20%

**-------------------------------------------------------------------------------------------------**

**Ouptut:**


El monto total a pagar por Alberto es 12000.

El monto total a pagar por Freddy es 9600.

El monto total a pagar por Yomira es 24000.

El monto total a pagar por Juan es 7200.
