# Con PyTorch

```python
import torch
```

```python
a = torch.tensor(1.0, requires_grad = True)
b = torch.tensor(2.0, requires_grad = True)
c = torch.tensor(3.0, requires_grad = True)
```

```python
d = a + b
e = b - c
```

```python
y = d * e
```

```python
y.backward()
```
