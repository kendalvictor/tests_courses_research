Una clase es como un molde, del que luego se pueden crear mltiples objetos.

```python
class Auto:
  "Este es un ejemplo de clase para crear objetos (autos)"

  # Este metodo inicializa los datos.
  # Es automaticamente invocado cuando el objeto de la clase es creado

  # self es usado como referencia del objeto de la clase.
  def __init__(self, empresa, modelo):
    self.empresa = empresa
    self.modelo = modelo
  

  def imprimir(self):
    print(f'Este auto es de la empresa {self.empresa} y su modelo es {self.modelo} ')
```

```python
# Accedemos al doc string
Auto.__doc__
```

```python
# Ac hacemos el objeto ferrari.
# Estamos creando un objeto de la clase Auto
ferrari = Auto("ferrari", 2020)
```

```python
ferrari.imprimir()
```

## Herencia

```python
class Felino:
  def __init__(self):
    print("Este felino est listo")

  def numero_de_patas(self):
    print("Este felino tiene 4 patas")
  
  def tipo_de_comida(self):
    print("Este felino gusta de todo tipo de carne")
```

```python
class Gato(Felino):
  def __init__ (self):
    super().__init__() # Con esta linea de codigo estamos "jalando" el print. // print ("Este felino est listo")
    print("Este gato est listo")

  def tamao_gato(self):
    print("Mido menos de un metro")
  
  def tipo_de_comida(self):
    print("Este gato solo come atn")
```

```python
zazie = Gato()
```

```python
zazie.tamao_gato()
```

```python
zazie.tipo_de_comida()
```

```python
zazie.numero_de_patas()
```

## Encapsulacin

```python
class Persona:
  def __init__(self, nombre, apellido, celular):
    self.nombre = nombre
    self.apellido = apellido
    self.__celular = celular
  
  def mostrar_datos(self):
    print(self.nombre)
    print(self.apellido)
    print(self.celular)
```

# Getter y Setter

```python
class Persona:
  def __init__(self, nombre, apellido, celular):
    self.nombre = nombre
    self.apellido = apellido
    self.__celular = celular
  
  def mostrar_datos(self):
    print(self.nombre)
    print(self.apellido)


  def getCelular(self):
    print(self.__celular)
  
  def setCelular(self, celular):
    self.__celular = celular
```

```python
ricardo = Persona("Ricardo", "Flores", 73070872)

ricardo.mostrar_datos()
```

```python
ricardo.getCelular()
```

```python
ricardo.setCelular(9394943)
```
