### ¬øQu√© es Numpy?

<b>Seg√∫n la documentaci√≥n:.- </b> NumPy es un paquete fundamental para la computaci√≥n cient√≠fica en Python. Es una biblioteca de Python que proporciona un multidimensional array object, varios objetos derivados y una variedad de rutinas para operaciones r√°pidas en arrays.

### ¬øPor qu√© usar Numpy?

*   Ofrece arrays que son mucho m√°s veloces y compactos que las listas Python. 
*   Consume menos memoria al momento de almacenar los datos.

```python
import time
size_of_vec = 10000

def pure_python_version():
    t1 = time.time()
    X = range(size_of_vec)
    Y = range(size_of_vec)
    Z = [X[i] + Y[i] for i in range(len(X)) ]

    return time.time() - t1

def numpy_version():
    t1 = time.time()
    X = np.arange(size_of_vec)
    Y = np.arange(size_of_vec)
    Z = X + Y
    return time.time() - t1


t1 = pure_python_version()
t2 = numpy_version()
print(t1, t2)
print("Numpy es " + str(t1/t2) + " m√°s veloz!")
#Cr√©ditos: webcourses.ucf.edu
```

## Multiplicaci√≥n de matrices
Captura de Pantalla 2021-05-21 a la(s) 21.25.17.png

```python
#Regla 1, Asociatividad
#.................A(BC)
r1_mul_1 = np.dot(matrix_A, np.dot(matrix_B, matrix_C))
print(r1_mul_1)
print("\n")
#.................(AB)C
r1_mul_2 = np.dot(np.dot(matrix_A, matrix_B), matrix_C)
print(r1_mul_2)
```

```python
#Regla 2, Distribuidad
#‚Ä¶................A(B+C)
r1_mul_1 = np.dot(matrix_A, (matrix_B + matrix_C))
print(r1_mul_1)
print("\n")
#.................AB + AC
r1_mul_2 = np.dot(matrix_A, matrix_B) + np.dot(matrix_A, matrix_C)
print(r1_mul_2)
```

```python
#Regla 3, Distribuidad
r1_mul_1 = np.dot((matrix_A + matrix_B), matrix_C)
print(r1_mul_1)
print("\n")
r1_mul_2 = np.dot(matrix_A, matrix_C) + np.dot(matrix_B, matrix_C)
print(r1_mul_2)
```

```python
#Regla 4, Asociatividad escalar
k = 2
#..................(kA)C
r1_mul_1 = np.dot(np.dot(k, matrix_A), matrix_C)
print(r1_mul_1)
print("\n")
#.................k(AC)
r1_mul_2 = np.dot(k, np.dot(matrix_A, matrix_C))
print(r1_mul_2)
```
