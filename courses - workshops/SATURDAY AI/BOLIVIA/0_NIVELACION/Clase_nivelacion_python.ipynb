### Una condicional más

```python
def operadora(opcion):
  '''
  opcion 1: contactarse con asistencia
  opcion 2: contactarse con servicio tecnico
  opcion 3: contactarse con el gerente
  '''
  if (opcion == 1):
    print("Se te contactara con asistencia")
  
  elif (opcion == 2):
    print("Se te contactara con servicio tecnico")
  
  elif (opcion == 3):
    print("Se te contactara con gerente")
  
  else:
    print("Hasta luego")
```

```python
operadora(9)
```
