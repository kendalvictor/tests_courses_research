## Prepare the data

```
x_train,  y_train = X, Y
```

##  Build the model

```
def linear_regression(x, y , alpha, epochs):
    w = np.random.rand(1)[0]
    b = np.random.rand(1)[0]

    Y_hat = w*x + b

    loss = []
    for i in range(epochs):

        loss.append( 0.5*np.sum((Y_hat - y)**2) )

        w = w - alpha*np.sum( (Y_hat - y)*x )
        Y_hat = w*x + b

        b = b - alpha*np.sum( (Y_hat - y) )
        Y_hat = w*x + b
        
        #print(i, loss[-1])

    return {
        'w':w,
        'b':b,
        'loss':loss
    }


    
```

## Train the model

```
result = linear_regression(x = x_train, y = y_train, alpha= 0.0001, epochs=10000 )
```

```
result['loss'][-1]
```

```
plt.plot(result['loss'], label='loss')
plt.legend()
```

## Evaluate the model

```
Y_hat = result['w']*X + result['b']
plt.scatter(X, Y)
plt.plot(X, Y_hat)
plt.xlabel('Average number of rooms')
plt.ylabel('House Prices')
plt.title('Boston house prices dataset')
```
