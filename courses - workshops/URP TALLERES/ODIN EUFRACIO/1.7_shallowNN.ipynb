## Make predictions

```
def predict( results, X ):

  A0 = X
  W1 = results['W1']
  b1 = results['b1']

  W2 = results['W2']
  b2 = results['b2']

  # Forward propagation
  Z1 = W1 @ A0 + b1
  A1 = np.tanh(Z1)

  Z2 = W2 @ A1 + b2
  A2 = ( 1.0 + np.exp(-Z2) )**(-1) 

  yhat = np.around(A2)
  return yhat
```
