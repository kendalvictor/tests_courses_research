##  Build the model

```
def linear_regression_GD(X, y , alpha, epochs):
    
    w = np.random.rand(X.shape[0], 1)
    b = np.random.rand(1)[0]
    m = X.shape[1]
    
    loss = []
    for i in range(epochs):
      
      z = w.T @ X + b
      yhat = z
      
      dz = yhat - y

      dw = (1.0/m) * ( X @ dz.T )
      db = (1.0/m) * np.sum(dz)

      w = w - alpha*dw
      b = b - alpha*db

      loss.append( (1.0/m)*np.sum((dz**2)) )  
      
    return {
        'w':w,
        'b':b,
        'loss':loss
        }
```
