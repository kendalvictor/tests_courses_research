## 5. Pre modelamiento

```python
#Identify the target and columns types

ID_col = ['TIP_DOC','NUM_DOC']
target_col = ['TRG_3M']
cat_cols = ['FLG_CLIENTE_IBK_CA','GENERO_F','GENERO_M']
other_col=['PERIODO'] #Test and Train Data set identifier
num_cols= list(set(list(tv_db.columns))-set(cat_cols)-set(ID_col)-set(target_col)-set(other_col))
```

```python
#Imputando datos

#Impute numerical missing values with 0
tv_db[num_cols] = tv_db[num_cols].fillna(0)
tv_db[cat_cols] = tv_db[cat_cols].fillna("-9999")
```

```python
#Transformando variables categoricas a string
number = LabelEncoder()
tv_db['FLG_CLIENTE_IBK_CA'] = number.fit_transform(tv_db['FLG_CLIENTE_IBK_CA'].astype('str'))
tv_db['GENERO_F'] = number.fit_transform(tv_db['GENERO_F'].astype('str'))
tv_db['GENERO_M'] = number.fit_transform(tv_db['GENERO_M'].astype('str'))
tv_db['TRG_3M'] = number.fit_transform(tv_db['TRG_3M'].astype('str'))
```

```python
#transformando categoricas a dummies
columnas_dummies = []
for col in cat_cols:
    dummy = pd.get_dummies(tv_db[col],prefix=col)
    columnas_dummies = columnas_dummies + list(dummy)
    tv_db = pd.concat([tv_db,dummy],axis=1)
```
