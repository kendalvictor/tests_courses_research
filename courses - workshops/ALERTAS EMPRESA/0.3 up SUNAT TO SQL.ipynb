```python
import pandas as pd
```

```python
csv_name = "cantidad_trabajadores.csv"
cant_trab = pd.read_csv(csv_name, dtype={'RUC': str, 'PERIODO': str})
print(cant_trab.shape)

cant_trab = cant_trab.drop_duplicates(subset=['PERIODO', 'RUC'], keep='last')
cant_trab.to_csv(csv_name, index=False)
print(cant_trab.shape)
```

```python
cant_trab.columns = [col.upper().strip() for col in cant_trab.columns]
```

```python
cant_trab.dtypes
```

```python
cant_trab['PERIODO'].value_counts(dropna=False)
```

```python
def get_nro_trab_num(val):
    pre = val.replace(' ', '').split('.')[0]
    
    if pre.isnumeric():
        return int(pre)
    
    try:
        return int(pre)
    except:
        print(pre, end=' ')
        return None
    
    return None


cant_trab['TOTAL_TRABAJADORES'] = cant_trab['TOTAL_TRABAJADORES'].apply(lambda _: get_nro_trab_num(_))
cant_trab['TOTAL_TRABAJADORES'].value_counts(dropna=False).sort_index()
```

```python
cant_trab['PENSIONISTA'] = cant_trab['PENSIONISTA'].apply(lambda _: get_nro_trab_num(_))
```

```python
cant_trab['PRESTADOR_SERVICIO'] = cant_trab['PRESTADOR_SERVICIO'].apply(lambda _: get_nro_trab_num(_))
```
