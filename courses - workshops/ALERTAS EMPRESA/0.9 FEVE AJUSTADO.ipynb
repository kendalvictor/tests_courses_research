```python
import glob
import pandas as pd

relacion = []
```

```python
for name in glob.iglob('FEVE_REAJUSTADO_BANCA_EMPRESA_*.xlsx'):
    relacion.append(
        (name, name.split('.')[0].split('_')[-1] )
    )
    print(relacion[-1])
```

```python
todo = []

for _excel, periodo in relacion:
    print("/"*100 , periodo)
    col_seguir_covid = 'SEGUIR COVID'
    
    # lectura de cada excel
    data = pd.read_excel(_excel, dtype={
        'CU': str, 'CLIENTE': str,  col_seguir_covid: str
    }, engine='openpyxl')
    
    # formato a las columnas y selccino las de interes
    data.columns = [_.upper() for _ in data.columns]
    data = data[['CU', 'CLIENTE', col_seguir_covid]]
    
    
    # formato homogeneo al conntenido, filtrado y seteado decolumnas de apoyo
    for col in data.columns:
        data[col] = data[col].fillna('').apply(lambda _: _.upper())
        
    data = data[data[col_seguir_covid] == 'SI']
    data['PERIODO'] = str(periodo)
    data['SEGUIR_COVID'] = 1
    
    # Acumulado de info con los dem√°s periodos
    print(data.shape)
    todo.append(data)
    display(data.head(3))
    
    del data
```

```python
data = pd.concat(
    todo, axis=0
)
data['PERIODO'].value_counts()
```
