```python
def conversion_dias(val):
    if pd.isnull(val):
        return '-'
    
    mensaje = '{} {} en días'.format(
        'aumenta' if val >= 0 else 'disminuye', int(np.abs(round(val, 0)))
    )
    return mensaje
```

```python
def conversion_porcentaje(val):
    if pd.isnull(val):
        return '-'
    
    mensaje = '{} {} %'.format(
        'aumenta' if val >= 0 else 'disminuye', int(np.abs(round(val*100, 0)))
    )
    return mensaje
```

```python
def conversion_porcentaje2(val):
    if pd.isnull(val):
        return '-'
    
    mensaje = '{} {} %'.format(
        'aumenta' if val >= 0 else 'disminuye', int(np.abs(round(val, 2)))
    )
    return mensaje
```

```python
def get_valores_periodo_cliente(df, funcion_filtro, cols_alerta_banca, cols_valores_all, col_alertado_banca):
    list_df = []
    df = funcion_filtro(df)
    df = df[df[col_alertado_banca] >= 1][['COD_UNICO', 'PERIODO'] + cols_alerta_banca + cols_valores]
    
    for orden_banca, alerta in enumerate(cols_alerta_banca, start=1):
        sub_alerta = df[df[alerta] == 1][['COD_UNICO', 'PERIODO'] + cols_valores]
        sub_alerta['CODIGO_ALERTA'] = alerta.split('_')[0]
        sub_alerta['ORDEN_BANCA'] = 'A{}'.format(orden_banca)
        sub_alerta['VALOR_V1'] = None
        sub_alerta['VALOR_V2'] = None
        sub_alerta['VALOR_V3'] = None
        
        for col_valor in [col for col in df.columns if col.startswith(alerta.replace("NEW", '')) and 'VALUE' in col.upper()]:
            if '_1_' in col_valor:
                sub_alerta['VALOR_V1'] = sub_alerta[col_valor].copy()
            if '_2_' in col_valor:
                sub_alerta['VALOR_V2'] = sub_alerta[col_valor].copy()
            if '_3_' in col_valor:
                sub_alerta['VALOR_V3'] = sub_alerta[col_valor].copy()
            
        sub_alerta = sub_alerta[['CODIGO_ALERTA', 'ORDEN_BANCA', 'COD_UNICO', 'PERIODO', 'VALOR_V1', 'VALOR_V2', 'VALOR_V3']]
        list_df.append(sub_alerta)
        
    df = pd.concat(list_df, axis=0, ignore_index=True)
    print(df.shape)
    
    return df
```

```python
def get_alertas_periodo(df, funcion_filtro, listas_alertas):
    llave = ['COD_UNICO', 'PERIODO']
    df = funcion_filtro(df)[llave + listas_alertas]
    
    df_periodo = df.melt(id_vars=llave, 
                        var_name="ALERTAS_ACTIVAS", 
                        value_name="flg_alert")
        
    df_periodo = df_periodo[df_periodo['flg_alert'] == 1].groupby(
            by=llave
    ).agg({'ALERTAS_ACTIVAS': list})
    
    df = df.merge(df_periodo, on =llave, how='left')[llave + ['ALERTAS_ACTIVAS']]
    del df_periodo
    
    return df
```

```python
def validation_mismas_alertas(activas, pasadas):
    if not(isinstance(activas, list)  and isinstance(pasadas, list) ):
        # -- entonces no llegaron a tener alerta
        return 0
    
    for alerta in activas:
        # si la alerta no la tenido el mes anterior  ooooooo   es una alerta alterada radicalmente retornar "0"
        if alerta not in pasadas or alerta in alertas_alteradas_a_no_considerar: 
            return 0
    return 1

def add_alertas_mes_anterior(df, df_anterior):
    df = df.merge(df_anterior, on=['COD_UNICO', 'PERIODO'], how='left')
    
    df['MISMAS_ALERTAS'] = df[['ALERTAS_ACTIVAS', 'ALERTAS_MES_ANTERIOR']].apply(lambda _: validation_mismas_alertas(_[0], _[1]), axis=1)
    return df
```
